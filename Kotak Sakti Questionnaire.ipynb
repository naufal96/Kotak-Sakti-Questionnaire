# Show the Query for number of customers purchasing more than 5 books
 

```python
import pandas as pd
```

```python
df1 = pd.read_csv('customers.csv')
df1
```

```python
df2 = pd.read_csv('invoices.csv')
df2
```

```python
df3 = pd.read_csv('invoice_lines.csv')
df3
```

```python
df4 = df3\
    .merge(df2, left_on='invoice_id', right_on='id', how='left')\
    .merge(df1, left_on='customer_id', right_on='id')
df4
```

```python
df5 = df4.groupby(['name'], as_index=False)['quantity'].sum()
df5
```

```python
df5.query('quantity >= 5')
```

# Show the SQL query for a list of customers who never purchased anything

```python
df6 = df1.merge(df5, on = 'name', how = 'left')
df6
```

```python
df7 = df6.fillna(0)
df7
```

```python
df7.query('quantity == 0')
```
